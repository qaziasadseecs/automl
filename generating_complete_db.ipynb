```python
import pandas as pd
from itertools import product

# Load datasets
df1 = df_merged.copy()
df2 = df_specifications.copy()

# Generate all combinations of rows from both datasets
combined_rows = list(product(df1.values, df2.values))
combined_data = pd.DataFrame(columns=list(df1.columns) + list(df2.columns))
# Populate combined data
for i, (row1, row2) in enumerate(combined_rows):
    combined_data.loc[i] = list(row1) + list(row2)
```

```python
combined_data.head()
```

```python
combined_data.columns
```

```python

# Define weights for each model
model_weights = {
    "Top_Model_1": 3,
    "Top_Model_2": 2,
    "Top_Model_3": 1,
    "M1": 3,
    "M2": 2.5,
    "M3": 2,
    "M4": 1.5,
    "M5": 1,
    "M6": 0.5
}

def calculate_combined_score(row):
    combined_score = {}
    for col in ["Top_Model_1", "Top_Model_2", "Top_Model_3", "M1", "M2", "M3", "M4", "M5", "M6"]:
        model = row[col]
        if model in combined_score:
            combined_score[model] += model_weights[col]
        else:
            combined_score[model] = model_weights[col]
    
    # Find the model with the highest combined score
    best_model = max(combined_score, key=combined_score.get)
    return best_model




# Apply the function to calculate the combined score
combined_data["Best_Model"] = combined_data.apply(calculate_combined_score, axis=1)
```

```python
df_final = combined_data.drop(columns=["Top_Model_1","Top_Model_2","Top_Model_3", "M1", "M2", "M3", "M4", "M5", "M6","Other Factors"])
```
