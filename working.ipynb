```python
import pandas as pd

# Read the performance.csv file into a DataFrame
df = pd.read_csv('Performance.csv')

# Create an empty DataFrame to store the top-3 models for each file
top_models_df = pd.DataFrame(columns=['Filename', 'Top_Model_1', 'Top_Model_2', 'Top_Model_3'])

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Get the filename
    filename = row['Filename']
    
    # Drop the filename column to only have model names and scores
    model_scores = row.drop('Filename')
    
    # Sort the models based on their scores and get the top-3 models
    top_models = model_scores.sort_values(ascending=False)[:3].index.tolist()
    
    # Add the filename and top-3 models to the DataFrame
    top_models_df = top_models_df.append({'Filename': filename,
                                          'Top_Model_1': top_models[0],
                                          'Top_Model_2': top_models[1],
                                          'Top_Model_3': top_models[2]}, ignore_index=True)

# Display the DataFrame
top_models_df.head(5)
```

```python
top_models_df.to_csv('results.csv', index=False)
```
